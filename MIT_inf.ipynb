### MODELING UNCERTAINTY
- to model uncertainty, need to track 
    1. sample space 
    2. probability of each possible outcome 

**sample space**: set of all possible outcomes of an xperiment (Ex: heads or tails), denoted by $\Omega$

**probability**: denoted by P (with two lines for p) ex: P(S) is probability of S happening

**collectively exhaustive**: every possible outcome is in it ($\omega$)

**mutually exhaustive**: exactly one possible outcome  

some rules:
- after experiment is run, mutually exhaustive outcome, can't be more than 1 or less than 0
- the probabilities of each outcome is between 0 and 1
- all probabilities sum up to 1


### Events
**event**: a subset of $\Omega$, 
subset of possible outcomes 
ex: subset of outcomes where second of 2 flips is tails 

- events can overlap w each other 
- probability of event happening = sum of all probabilities of the event's outcomes 

Events as sets

There are two special events in every model 
- $\Omega$ itself, the entire set 
- $\phi$, empty set and complement of $\Omega$

properties of these events:
- $\Omega$ has probability of 1, $\phi$ has probability of 0
- if an event is a subset of the other event, then its probability has to be less than the probability of the other event by definition
    - given that every other event that is not $\Omega$ is a subset of it, therefore its probability is less than 1 
- 2 events that don't overlap have an empty set intersection 

```python
# function that finds probability of event, which is sum of probability of alloutcomes in the event

def prob_of_event(event, prob_space):
    total = 0
    for outcome in event:
        total += prob_space[outcome]
    return total

prob_space = {'sunny': 1/2, 'rainy': 1/6, 'snowy': 1/3}
rainy_or_snowy_event = {'rainy', 'snowy'}
print(prob_of_event(rainy_or_snowy_event, prob_space))
```

### Random Variables 

Example of simulating weather for tomorrow:
- you'll notice that the weather and I (indicator) will change randomly, perhaps at rate indicated in the probability space
- this is us sampling from the probability space, which gets us a random outcome $\omega$

```python
random_outcome = comp_prob_inference.sample_from_finite_probability_space(prob_space)
W = random_outcome
if random_outcome == 'sunny':
    I = 1
else:
    I = 0

print(W)
print(I)
```

- random variable: given finity probability space, the ran var. maps from $\Omega$ to diff set of values that the ran var can take on
- these random variables can map the set to a different set, like in the definition of functions 
    - ex: I mapping {sunny,rainy,snowy} to {1,0}


```python
W_mapping = {'sunny': 'sunny', 'rainy': 'rainy', 'snowy': 'snowy'}
I_mapping = {'sunny': 1, 'rainy': 0, 'snowy': 0}

random_outcome = comp_prob_inference.sample_from_finite_probability_space(prob_space)
W = W_mapping[random_outcome]
I = I_mapping[random_outcome]

print(W)
print(I)
```

probability mass function: aka probability table and probability distribution 
- the probability of an event of a finite probability space happening Px(x) 
- small x (outcome) belonging to big X (a certain event)
- X denoting random values, x denoting nonrandom values 

ex: pw(rainy) = 1/6 or pi(1)= 1/2
- remember that all entries fall between probability value of 0 and 1 
- in other words, summation of all entries px(x) = 1
